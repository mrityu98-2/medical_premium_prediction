```python
bin_labels = ['Low', 'Moderate Low', 'Moderate', 'Moderate High', 'High']
df['PremiumLabel']=pd.cut(df['PremiumPrice'],bins=5,labels=bin_labels,precision=0)
df['AgeLabel']=pd.cut(df['Age'],bins=5,labels=bin_labels,precision=0)
df['WeightLabel']=pd.cut(df['Weight'],bins=5,labels=bin_labels,precision=0)
df['HeightLabel']=pd.cut(df['Height'],bins=5,labels=bin_labels,precision=0)
```

```python
df_ohe=pd.get_dummies(data=df,columns=['PremiumLabel','AgeLabel','WeightLabel','HeightLabel'],drop_first=True)
df_ohe.sample(4)
```

```python
y=df_ohe['PremiumPrice']
x=df_ohe.drop(columns=['PremiumPrice'],axis=1)
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=33,test_size=0.3)
```

```python
from sklearn.feature_selection import mutual_info_regression
mi_scores = mutual_info_regression(x_train,y_train)
mi_scores = pd.Series(mi_scores, name="MI Scores",index=x.columns)
mi_scores = mi_scores.sort_values(ascending=False)
(mi_scores*100).head(15).index
```

```python
feature_list=list((mi_scores*100).head(15).index)
x_train=x_train[feature_list]
x_test=x_test[feature_list]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_train_scaled.shape
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_test_scaled.shape
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
rf_reg=RandomForestRegressor(n_estimators=20,random_state=42)
rf_reg.fit(x_train_scaled,y_train)
```

```python
rf_reg_pred=rf_reg.predict(x_test_scaled)
rf_reg_accuracy=r2_score(y_test,rf_reg_pred)
print(f'Accuarcy is sort of {rf_reg_accuracy*100:.2f}%')
```

```python
rf_reg_mse = mean_squared_error(y_test,rf_reg_pred)
print(f'Mean Squared Error: {rf_reg_mse:.2f}')
```

```python
rf_reg_mae = mean_absolute_error(y_test,rf_reg_pred)
print(f'Mean Absolute Error: {rf_reg_mae:.2f}')
```

```python
pickle.dump(rf_reg, open('rf_reg_insurance.pkl', 'wb'))
```

```python
plt.figure(figsize = (8,8))
sb.displot(y_test-rf_reg_pred,kde=True)
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.scatter(y_test,rf_reg_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
param_grid = {
    'n_estimators': [10,20,25,30,40,50,60,65,85,100,120],
    'max_features': ['sqrt','log2',1.0],
    'max_depth': [6,7,9,11,12],
    'criterion':['squared_error','absolute_error']
} 
```

```python
rf_reg_hyp=GridSearchCV(RandomForestRegressor(random_state=42),param_grid=param_grid,cv=5,n_jobs=-1,verbose=1,scoring='neg_mean_squared_error')
rf_reg_hyp.fit(x_train_scaled,y_train)
```

```python
best_params = rf_reg_hyp.best_params_
print(f'Best parameters: {best_params}')
```

```python
rf_reg_clf = rf_reg_hyp.best_estimator_
rf_reg_hyp_pred = rf_reg_clf.predict(x_test_scaled)
```

```python
rf_reg_hyp_accuracy=r2_score(y_test,rf_reg_hyp_pred)
print(f'Accuarcy is sort of {rf_reg_hyp_accuracy*100:.2f}%')
```

```python
rf_reg_hyp_mse = mean_squared_error(y_test,rf_reg_hyp_pred)
print(f'Mean Squared Error: {rf_reg_hyp_mse:.2f}')
rf_reg_hyp_mae = mean_absolute_error(y_test,rf_reg_hyp_pred)
print(f'Mean Absolute Error: {rf_reg_hyp_mae:.2f}')
```

```python
gb_reg = GradientBoostingRegressor(random_state=42)
gb_reg.fit(x_train_scaled,y_train)
```

```python
gb_reg_pred=gb_reg.predict(x_test_scaled)
```

```python
gb_reg_accuracy=r2_score(y_test,gb_reg_pred)
print(f'Accuarcy is sort of {gb_reg_accuracy*100:.2f}%')
gb_reg_mse = mean_squared_error(y_test,gb_reg_pred)
print(f'Mean Squared Error: {gb_reg_mse:.2f}')
gb_reg_mae = mean_absolute_error(y_test,gb_reg_pred)
print(f'Mean Absolute Error: {gb_reg_mae:.2f}')
```
